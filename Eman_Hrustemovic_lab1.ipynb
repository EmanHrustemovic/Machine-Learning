### 1.   Read and Write operations on a file

In order to analyze and work with our dataset, firstly, we need to read its content. You can use the following code snippet. Additionally, you can use the helper *is_number* function, to handle possible non-numeric values in the columns.

````python
def is_number(s):
    if s is None:
        return False
    try:
        float(s)
        return True
    except (ValueError, TypeError):
        try:
            int(s)
            return True
        except (ValueError, TypeError):
            return False
def read_file(file_path):
    try:
        with open(file_path, 'r') as file:
            data = []
            lines = file.readlines()

            keys = lines[0].split(',')
            for i in range(1, len(lines)):
                data_row = [
                    float(cell.strip('\n')) if is_number(cell.strip('\n')) else cell.strip('\n')
                    for cell in lines[i].split(',')
                ]
                data.append(data_row)
            return data
    except FileNotFoundError:
        raise RuntimeError("File not found.")
    except Exception as e:
        raise RuntimeError(f"An error occurred: {e}")
````

#### Standard Deviation



> Standard deviation is a statistic that measures the dispersion of a dataset relative to its mean and is calculated as the square root of the variance. To calculate the standard deviation, we firstly divide the squared difference of the column and its mean, by the total number of values. After that, we take the square root and obtain the $ \sigma  $ value.

> $ \sigma = \sqrt{\frac{\sum_{i=1}^{N} (x_i - \mu)^2}{N}} $








````python
def std_dev(data, index):
  pass
````

````python
def std_dev(data, index):
    data = [row[index] for row in data if row[index] is not None]

    n = len(data)

    if n == 0:
        raise ValueError("No valid values found in the specified column.")

    mean_value = sum(data) / n
    squared_diff_sum = sum((x - mean_value) ** 2 for x in data)

    # If n is 1, standard deviation is undefined, return 0
    if n == 1:
        return 0

    std_deviation = (squared_diff_sum / (n - 1)) ** 0.5
    return std_deviation

# Example usage
colum1 = [5.1, 4.9, 4.7, 4.6, 5.0, 5.4, 4.6, 5.0, 4.4, 4.9, 5.4, 4.8, 4.8, 4.3, 5.8, 5.7, 5.4]

std_dev_colum1 = std_dev(iris_data, 0)
print(f"Standard Deviation Colum1: {std_dev_colum1}")
````

#### Variance

> Variance is a statistical measurement used to determine how far each number is from the mean and from every other number in the set.

> $ \text{Variance} = \sigma^2 $



````python
def variance(data, index):
  pass
````

````python
def variance(data, index):
    data = [row[index] for row in data if row[index] is not None]

    n = len(data)

    if n == 0:
        raise ValueError("No valid values found in the specified column.")

    mean_value = sum(data) / n
    squared_diff_sum = sum((x - mean_value) ** 2 for x in data)

    # If n is 1, variance is undefined, return 0
    if n == 1:
        return 0

    variance_value = squared_diff_sum / (n - 1)
    return variance_value


iris_data = [
    [5.1, 3.5, 1.4, 0.2, 'Iris-setosa'],
    [4.9, 3.0, 1.4, 0.2, 'Iris-setosa'],
    # ... (other rows)
]

# Assuming you want to calculate the variance for SepalLengthCm (index 0)
variance_sepal_length = variance(iris_data, 0)
print(f"Variance Sepal Length: {variance_sepal_length}")
````

````python
import matplotlib.pyplot as plt
import numpy as np

def plot(labels, values, type='hist'):
    if type == 'hist':
        plt.bar(labels, values)
        plt.xlabel('Values')
        plt.ylabel('Frequency')
        plt.title('Histogram')
        plt.show()
    elif type == 'pie':
        plt.pie(values, labels=labels, autopct='%1.1f%%')
        plt.title('Pie Chart')
        plt.show()
    else:
        raise ValueError("Invalid plot type. Supported types are 'hist' and 'pie'.")

# Example usage
colum1 = [5.1, 4.9, 4.7, 4.6, 5.0, 5.4, 4.6, 5.0, 4.4, 4.9, 5.4, 4.8, 4.8, 4.3, 5.8, 5.7, 5.4]

# Plot histogram before normalization
plot('Values', colum1, type='hist')

# Normalize colum1
normalized_colum1 = [(x - min(colum1)) / (max(colum1) - min(colum1)) for x in colum1]

# Plot histogram after normalization
plot('Values', normalized_colum1, type='hist')

# Plot pie chart for Species column
species_labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
species_values = [50, 50, 50]

plot(species_labels, species_values, type='pie')
````
